## Pandas 공식문서 튜토리얼 따라하기

# 2장 표 형식 데이터를 어떻게 읽고 쓰나요?

```python
import pandas as pd
```

```python
# 타이타닉 데이터셋 불러오기
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic # 891개의 행과 15개의 열(특징)을 가졌다.
```

```python
#처음 행 8개 확인하기
titanic.head(8) # 처음 n행을 볼려면 head(n) 메소드를 활용하면 된다.
```

```python
# 끝의 행 확인하기
titanic.tail(8) # tail(n) 끝의 n행을 확인한다.
```

```python
# pandas 열의 유형을 확인하기
# dtypes를 활용한다. dtypes는 속성이기에 괄호를 쓰지 않는다.
titanic.dtypes
```

# 타이타닉 데이터를 excel로 요청

```python
titanic.to_excel('titanic.xlsx', sheet_name="sheet1", index=False) # 데이터 셋을 excel로 저장
titanic = pd.read_excel('titanic.xlsx', sheet_name='sheet1') # excel로 저장된 데이터 셋 읽어오기
titanic.head()
```
