# 8장 여러 테이블의 데이터를 결합

```python
import pandas as pd
air_quality_no2 = pd.read_csv("/content/air_quality_no2_long (1).csv", parse_dates=True)
air_quality_pm25 = pd.read_csv("/content/air_quality_pm25_long.csv", parse_dates=True)
```

```python
air_quality_no2
```

```python
air_quality_pm25
```

```python
# 공통 column으로 테이블 맞추기
air_quality_no2 = air_quality_no2[["date.utc", "location","parameter","value"]]
air_quality_no2.head()
```

```python
# 공통 column으로 테이블 맞추기
air_quality_pm25 = air_quality_pm25[["date.utc","location", "parameter", "value"]]
air_quality_pm25.head()
```

# 객체연결

```python
# no2 데이터 와 pm25 데이터 테이블을 하나의 테이블로 결합
#axis = 0 은 row를 의미 axis = 1은 column을 의미
# concat()함수는 행 또는 열 중 하나를 따라 여러 테이블의 연결 작업을 수행
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)
```

```python
air_quality.head()
```

```python
air_quality.shape
```

```python
# 날짜/시간 정보에 따라 테이블을 정렬하면 parameter 정보를 통해 두 테이블의 조합을 알수있다.
air_quality = air_quality.sort_values(["date.utc"]) # date.utc 기준으로 내림차순
```

```python
# concat 사용하여 연결
air_quality_ = pd.concat([air_quality_pm25, air_quality_no2], keys = ["pm25", "no2"])
```

```python
air_quality_.head()
```

# 공통 식별자를 사용하여 테이블 조인

```python
# air_quality_stations 데이터셋 읽기
stations_quality = pd.read_csv("/content/air_quality_stations.csv")
```

```python
stations_quality.head()
```

```python
air_quality.head()
```

```python
# merge()기능을 사용하여 테이블의 각 행에 대해 air_quality 는 station_quality에 추가된다.
# merge()기능은 데이터베이스 운영과 유사한 여러 옵션을 지원
air_quality = pd.merge(air_quality, stations_quality, how="left", on="location")
```

```python
air_quality.head()
```

```python
parameters_quality = pd.read_csv("/content/air_quality_parameters.csv")
parameters_quality.head()
```

```python
# merge를 이용해 air_quality와 parameters_quality 결합하기
air_quality = pd.merge(air_quality, parameters_quality,
                       how="left", left_on="parameter", right_on="id")
air_quality.head()
```
