## Pandas 공식문서 튜토리얼 따라하기

# 6장 요약 통계 계산 방법

```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
# mean() 데이터열의 평균값을 출력
titanic['age'].mean()
```

```python
# 타이타닉 승객의 평균 연령과 티켓 가격 알아보기
titanic[['age','fare']].median() #median()함수는 열의 중앙값을 출력
```

```python
# describe()를 활용하여 age, fare의 통계를 알아보자
titanic[['age','fare']].describe()
```

# 카테고리별로 그룹화된 통계 집계

```python
titanic[["sex", "age"]].groupby("sex").mean()
```

```python
# sex, pclass 그룹화에 티켓가격 평균
titanic.groupby(["sex","pclass"])["fare"].mean()
```

# 카테고리별 레코드 수 계산

```python
# value_counts()함수는 열의 각 범주에 대한 값을 계산
titanic["pclass"].value_counts()
```
