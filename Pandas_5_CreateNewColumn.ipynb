## pandas 공식문서 튜토리얼 따라하기

# 기존 열에서 파생된 새 열을 만드는 방법

```python
import pandas as pd
air_qulity = pd.read_csv("/content/air_quality_no2.csv", index_col=0, parse_dates=True)
air_qulity.head()
```

```python
# 런던에 있는 관측소의 농도 확인하기
# 섭씨 25도, 압력을 1013hPa로 가정하면 환산계수는 1,882
# 런던 * 1,882
air_qulity["london_mg"] = air_qulity["station_london"] * 1882
air_qulity.head()
```

```python
# 두 열을 이용하여 새 열 만들기
# 파리와 엔드위프의 값을 나누고 결과값을 새로운 열로 만들기
air_qulity["ratio_paris_antwerp"] = air_qulity["station_paris"] / air_qulity["station_antwerp"]
air_qulity.head()
```

```python
# rename()을 활용하여 열의 이름을 변경
air_qulity.rename(
    columns={
        "station_london" : "런던이에요",
        "station_paris" : "파리에요",
        "station_antwerp" : "엔드워프에요",
    }
)
```
