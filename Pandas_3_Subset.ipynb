# 특정 행을 필터링 하기 위해서는

```python
# 35세 이상의 승객을 출력
age_35 = titanic[titanic['age'] > 35]
age_35['age']
```

```python
# 35세 이상의 크기
age_35.shape
```

```python
# 객실 클래스 2 와 3의 타이타닉 승객들을 출력
#isin() Series의 각 요소가 전달된 값 시퀀스의 요소와 정확히 일치하는지 여부를 보여준다
#boolean으로 반환
class_23 = titanic[titanic['pclass'].isin([2, 3])]
class_23.head()
```

```python
# age에 NaN 값을 필터링한다
# notna 값이 Not Null 값은 True로 매핑한다.
age_no_na = titanic[titanic['age'].notna()]
age_no_na.head()
```

```python
# 크기를 통해 변화가 일어난걸 확인할 수 있다.
age_no_na.shape
```

# 특정 행과 열을 어떻게 선택할까

```python
# 35세 이상의 승객 성별을 알아보자
# loc/iloc 연산자는 선택 괄호 [] 앞에 필요하다. loc/iloc을 사용할 때 쉼표 앞에 있는 부분은
# 원하는 행이고 쉼표 뒤에 있는 부분은 선택할려는 열이다.
adult_names = titanic.loc[titanic["age"] > 35, "who"]  #loc[행인덱싱, 열인덱싱]
adult_names.head()
```

```python
# 10 ~ 25행과 3 ~ 5열을 선택해보자
titanic.iloc[10:25, 3:5]  #iloc[행 인덱스, 열 인덱스] 숫자로 접근
```

```python
# iloc을 이용해 새 값에 할당
titanic.iloc[0:3, 3] = "anoymous"
titanic.head()
```
